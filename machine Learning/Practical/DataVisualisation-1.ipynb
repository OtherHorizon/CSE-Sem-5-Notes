```python
import matplotlib.pyplot as plt
import numpy as np

# define data values
x = np.array([1,2,3,4,5]) # X-axis points
y =  x*2 # Y-axis points

plt.plot(x,y) # Plot the chart
plt.show() # display
```

## Label the axis and the graph itself

```python
x = np.array([1,2,3,4,5]) 
y =  x*2 

plt.plot(x,y) 
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My First Line Graph")
plt.show()
```

## Create Multiple Graphs at the same time

```python
x = np.array([1,2,3,4,5]) 
y =  x*2 

plt.plot(x,y) 
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("First Graph")
plt.show()

plt.figure()
x1 = [2,4,6,8]
y1 = [3,5,7,9]
plt.plot(x1,y1,'-.') # the 3rd arg defines the style of the line
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Second Graph")
plt.show()
```

|Character|Definition|
|---------|----------|
|-.| dotted lines with . in between|
|-| dashed line|
|:|dotten line|

## Applying colors

```python
x = np.array([1,2,3,4,5]) 
y =  x*2 

plt.plot(x,y) 

x1 = [2,4,6,8]
y1 = [3,5,7,9]
plt.plot(x,y1,'-.') 
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Second Graph")
# fill color in betweeen the two lines
plt.fill_between(x,y,y1,color='red',alpha=0.5)
plt.show()

```
