```python
# From Debashish Nath 20BCS7965
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
```

```python
db = datasets.load_diabetes()
print(db)
```

```python
print(db.data)
```

```python
print(db.target)
```

```python
db_x =  db.data[:, np.newaxis, 2]  # slice the data to get only one feature i.e. the 2nd column using numpy
print(db_x)
```

```python
db_x_train = db_x[:-80] # training data is all the data except the last 80 rows
db_x_test = db_x[-80:] # test data is the last 80 rows

db_y_train = db.target[:-80]
db_y_test = db.target[-80:]
```

```python
model = linear_model.LinearRegression()
model.fit(db_x_train, db_y_train)  # fit the model to the training data
db_y_predicted = model.predict(db_x_test)
print(f'MSE : {mean_squared_error(db_y_test, db_y_predicted)}') # Mean Squared Error
print(f'Weight : {model.coef_}') # slope of the line
print(f'Intercept : {model.intercept_}') # y-intercept of the line
```

```python
plt.scatter(db_x_test,db_y_test)
plt.plot(db_x_test, db_y_predicted, color='red')
plt.show()
```
