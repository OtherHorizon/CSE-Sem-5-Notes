Here we imported the _`pandas`_ library and then initialized a variable called _`ds`_ that read the csv file that contains the dataset and then store the dataset into the variable

```python
import pandas as pd
ds =  pd.read_csv("../Datasets/Sample5 - Brest Cancer.csv")
```

_`.head(n)`_ function is used to display the first _`n`_ rows from the Dataframe

```python
ds.head()
```

_`.shape`_ is used to display the number of rows and columns of the Dataframe

```python
ds.shape
```

_`.info()`_ is used to get an overall information about the dataframe like the column names, data type, number of entires, etc

```python
ds.info()
```

_`.describe()`_ is used to see the mean, std, min, max, etc. _i.e._ a description of the Dataframe

```python
ds.describe()
```

_`.isnull()`_ returns if a cell in the Dataframe is empty or not

```python
ds.isnull()
```

_`.sum()`_ as the name suggest returns the sum

```python
ds.isnull().sum()
```

_`.mean()`_ is used to find the mean of a given column

```python
ds["Clump Thickness"].mean()
```
