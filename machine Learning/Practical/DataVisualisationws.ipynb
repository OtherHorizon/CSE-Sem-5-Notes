imported _`matplotlib.pyplot`_ for plot and _`numpy`_ for numerical computation.

```python
import matplotlib.pyplot as plt
import numpy as np
```

_`.plot(x,y)`_ plots the data points in the array _`x`_ and _`y`_.
and _`.show()`_ displays the plot.

```python
x = np.array([1,2,3,4,5]) # X-axis points
y =  x*2 # Y-axis points

plt.plot(x,y) # Plot the chart
plt.show()
```

Label the axis and the graph itself

```python
x = np.array([1,2,3,4,5]) 
y =  x*2 

plt.plot(x,y) 
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("A line graph")
plt.show()
```

Fill color

```python
x = np.array([1,2,3,4,5]) 
y =  x*2 

plt.plot(x,y) 

x1 = [2,4,6,8]
y1 = [3,5,7,9]
plt.plot(x1,y1,'-.') 
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Second Graph")
plt.fill_between(x1,y1,y,color='red',alpha=0.5)
plt.show()
```

Multiple plots at once

```python
plt.subplot(3,1,1) # 3 rows, 1 column, 1st subplot
plt.subplot(3,1,2) # 3 rows, 1 column, 2nd subplot
plt.subplot(3,1,3) # 3 rows, 1 column, 3rd subplot
```

```python
x = np.array([1,2,3,4,5]) 
y =  x*2 

plt.plot(x,y) 
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("First Graph")
plt.show()

plt.figure()
x1 = [2,4,6,8]
y1 = [3,5,7,9]
plt.plot(x1,y1,':') # the 3rd arg defines the style of the line
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Second Graph")
plt.show()
```

```python
x = np.arange(0,10,0.05)

plt.subplot(4,1,1) # 4 rows, 1 column, 1st subplot
plt.plot(x,np.sin(x)) # y = np.sin(x)

plt.subplot(4,1,2) # 4 rows, 1 column, 2nd subplot
plt.plot(x,np.cos(x))

plt.subplot(4,1,3) # 4 rows, 1 column, 3rd subplot
plt.plot(x,np.tan(x))

plt.subplot(4,1,4) # 4 rows, 1 column, 4rd subplot
plt.plot(x,x)

plt.show()
```

```python
x = np.arange(0,2,0.01)
y = 1 + np.sin(2*np.pi*x)

f,((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(3,2)
ax1.plot(x,y,color = "orange")
ax2.plot(x,y,color = "green")
ax3.plot(x,y,color = "red")
ax4.plot(x,y,color = "blue")
ax5.plot(x,y,color = "magenta")
ax6.plot(x,y,color = "black")

plt.show()
```

```python
import random as random
students = ["D","E","B","A"]
marks=[]

for i in range(0,len(students)):
    marks.append(random.randint(0,101))

plt.xlabel("Student's Name")
plt.ylabel("Marks Obtained")
plt.title("CLASS RECORDS")
plt.plot(students,marks,color="black",linestyle="-.", marker="*",markerfacecolor="red",markersize=12)
```
